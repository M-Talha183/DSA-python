# singly linked List 

```python
# constructor to initialize a new node with data
class Node :
    def __init__(self, data):
        self.data = data 
        self.next = None 
# Create the first node (head of the list)
Head = Node(10)
# Link the second node
Head.next = Node(20)
# Link the third node
Head.next.next = Node(30)
# Link the fourth node
Head.next.next.next = Node(40)

temp = Head 
# printing linked list
while temp is not None :
    print(temp.data , "  Next == >" , temp.next)
    temp = temp.next
```

# Traversal of Singly Linked List

```python
# a linked list node

class Node :
    def __init__(self , data) :
        self.data = data
        self.next = None
      
 # function to traverse and print the singly linked list
  
def print_linked_list (head):
    while head is not None :
        print(head.data , end=" ")
        
        if head.next is not None :
            print ("-->", end=" ")
        head = head.next
        
head = Node (10)
head.next = Node (20)
head.next.next = Node (30)  
head.next.next.next = Node (40)

print_linked_list(head)
        
```

# Insert a node at a specific position in a linked list


```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def insert_specific(head, new_data, position):

    if position == 0:
        new_node = Node(new_data)
        new_node.next = head
        return new_node

    curr = head
    for i in range(1, position-1):
        if curr is None:
            return "Position out of bounds"
        curr = curr.next
    print(curr.data,"current value before insertion")
    new_node = Node(new_data)
    # print(new_node.data, "new node data")
    new_node.next = curr.next
    # print(new_node.next.data,"new node next pointer")
    curr.next = new_node
    # print(curr.next.data, "current node next data after insertion")
    return head


def print_linked_list(head):
    while head:
        print(head.data, end="")
        if head.next:
            print(" --> ", end="")
        head = head.next
    print()


# Create list
head = Node(10)
head.next = Node(20)    
head.next.next = Node(30)
head.next.next.next = Node(40)
head.next.next.next.next = Node(50)

print_linked_list(head)

# INSERT AND UPDATE HEAD
head = insert_specific(head, 25, 3)

print_linked_list(head)
```
